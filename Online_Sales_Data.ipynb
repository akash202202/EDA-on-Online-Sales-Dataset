```python
def Outlier1(x):
  out=[]
  Q1=13.250000
  Q3=17.750000
  IQR=Q3-Q1
  upper=Q3+(1.5*IQR)
  lower=Q1-(1.5*IQR)
  for i in x:
    if (i>upper) or (i<lower):
      out.append(i)
    return out
```

```python
def find_outliers_IQR(dataset):
    q1=dataset.quantile(0.25)
    q3=dataset.quantile(0.75)
    IQR = q3-q1
    outliers = dataset[((dataset<(q1-1.5*IQR)) | (dataset>(q3+1.5*IQR)))]
    return outliers
```

```python
outliers = find_outliers_IQR(Data['Unit Price'])
print('Number of outliers: '+ str(len(outliers)))
print('Max outlier value: '+str(outliers.max()))
print('Min outlier value: '+str(outliers.min()))
```

```python
def non_outliers_IQR(dataset):
    q1=dataset.quantile(0.25)
    q3=dataset.quantile(0.75)
    IQR = q3-q1
    print(IQR)
    not_outliers = dataset[~((dataset<(q1-1.5*IQR)) | (dataset>(q3+1.5*IQR)))]
    return not_outliers
```

```python
not_outliers = non_outliers_IQR(Data['Unit Price'])
print('Number of non-outliers: '+ str(len(not_outliers)))
print('Max non-outlier value: '+str(not_outliers.max()))
print('Min non-outlier value: '+str(not_outliers.min()))
```
